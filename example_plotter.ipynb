```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams['figure.dpi'] = 150

# Generate 11 random points
np.random.seed(43)  # Fix seed for reproducibility
# x = np.linspace(0.5, 1.5, 11)  # X-axis range: 0.5 to 1.5
# y = [8.07, 8.08, 8.10, 8.15, 8.09, 8.10, 8.08, 8.05, 8.04, 8.04, 8.07]

x= [0.5,0.6,0.7,0.8]
y= [7.92,7.96,8.11,8.14]

# Generate random confidence interval
y_err = np.random.uniform(0.05, 0.1, 4)  # Random error for confidence interval

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size

# 
plt.plot(x, y, color='#158BB8', marker='x', markersize=8, linestyle='--', linewidth=2,  label='Data Points', 
         markerfacecolor='k', markeredgecolor='k', markeredgewidth=2)

# 
plt.fill_between(x, y - y_err, y + y_err, color='#93CBFF', alpha=0.3, label='Confidence Interval')

#  X  Y 
# plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)

# 
light_gray = '#d1c2d3'  # 
plt.grid(True, linestyle='--', alpha=0.6)

# 
# plt.legend(fontsize=12)

# 
# plt.xlim(4.5, 15.5)
plt.ylim(6, 8.4)


plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

# light_purple = '#d0dfe6'
# plt.gca().set_facecolor(light_purple)


# 
for xi, yi in zip(x, y):
    plt.annotate(f'{yi:.2f}', (xi, yi), textcoords="offset points",
                 xytext=(0, 10), ha='center', fontsize=9)

# 
plt.tight_layout()

# 
plt.show()
```
